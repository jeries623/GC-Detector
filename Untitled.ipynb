```python
#import Statment
import pandas as pd
import numpy as np
import os

from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

```

```python
# Load Data
training = pd.read_csv('./data/dataa.csv')
training.head()
```

```python
# train a C45 decision tree and save the tree as an XML file
model = DecisionTreeClassifier(min_samples_split=2)

#pyC45.train(training_obs, training_cat, "DecisionTree.xml")
```

```python
X1=training.iloc[:,:-1]
Y1=training["Result"]
X1
```

```python
df=model.fit(X1, Y1)
df
```

```python
predict=df.predict(X1)
predict
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(predict,Y1)
```
